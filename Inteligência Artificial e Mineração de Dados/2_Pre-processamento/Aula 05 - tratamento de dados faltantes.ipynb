```python
import pandas as pd

data  = pd.read_csv('credito.csv')
data
```

```python
pd.isnull(data['age'])
```

```python
# para localizar os dados com problemas

data.loc[pd.isnull(data['age'])]
```

```python
# exclui a linha com valor null

data.dropna()
```

```python
#preenche com a média

data.fillna(data['age'].mean())
```

```python
# importa
import numpy as np
from sklearn.impute import SimpleImputer

#instancia o objeto
imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')

# define o array
X = [[7, 2, 3], [4, np.nan, 6], [10, 5, 9]]

# coloca o array dentro do objeto
imp_mean.fit(X)

# faz a transformação
print(imp_mean.transform(X))
```
