# Introdução

Nesse notebook vamos aprender como criar gráficos de linha usando o pandas e o matplotlib. Abandone definitivamente o excel e vamos mergulhar nessa ferramenta de visualização incrível.

Para esse tutorial irei utilizar o dataset "dates" disponibilizado exclusivamente pelo computersciencemaster com o tráfego mensal de um website. Portanto, vamos importar de dentro da pasta "data" o csv chamado "dates". A seguir vamos chamar o comando dates.head() para mostrar as primeiras 5 linhas do nosso dataset.

```python
import pandas as pd

dates = pd.read_csv("data/dates.csv")
dates['Date'] = pd.to_datetime(dates['Date'])
dates.head()
```

## Criando nosso primeiro gráfico de linhas

Agora vamos criar nosso primeiro gráfico de linhas simples utilizando algumas opções básicas do matplotlib. Mas antes, vamos importar a biblioteca usando o comando de importação.

```python
import matplotlib.pyplot as plt
```

A seguir, vamos criar um gráfico de linha usando apenas a quantidade de clicks e data em que aquela qunatidade de clicks foi registrada no website.

```python
plt.plot(dates["Date"], dates["Clicks"])
```

Agora que nosso primeiro gráfico de linhas está pronto, podemos perceber que existem alguns problemas. O primeiro deles é que nosso gráfico parece muito pequeno, sendo assim, vamos aumentá-lo usando o comando figsize.

```python
plt.figure(figsize=(10,6))
plt.plot(dates["Date"], dates["Clicks"])
plt.show()
```

Um outro problema que percebemos logo de cara é que não existem títulos para nada, nem para os eixos, nem para o gráfico, nem para as linhas. Certo, tudo isso faz parte da legibilidade de um gráfico, sendo assim, vamos dar uma melhorada começando pelo título do gráfico. Para isso vamos executar o seguinte comando:

```python
plt.figure(figsize=(10,6))
plt.plot(dates["Date"], dates["Clicks"])
plt.title("Clickes por data")
plt.show()
```

É importante também perceber que nossa ferramenta permite que você crie objetos ao plotar seus gráficos. Isso significa que ao criar uma "figure" você pode armazenar esse objeto em uma variável e manipular esse objeto posteriormente. Veja o exemplo:

```python
lc = plt.figure(figsize=(10,6))
lc = lc.add_axes([0,0,1,1])
```

Acima você pode ver um plano cartesiano completamente vazio e você só precisa "inserir" os dados no gráfico. Isso pode ser feito pelo comando plot já visto anteriormente. 

Certo, agora que já aprendemos usar essas variáveis em nossos gráficos, vamos adicionar um label para nossos eixos X e Y.

```python
lc = plt.figure(figsize=(10,6))

eixos = lc.add_axes([0,0,1,1])
eixos.plot(dates["Date"], dates["Clicks"])
eixos.set_ylabel("Clicks", fontsize=15)
eixos.set_xlabel("Datas de acesso", fontsize=15)
plt.title("Clicks por data", fontsize=20)
plt.show()
```

é possível também adicionar legendas ao nosso gráfico, usando o comando plt.legend:

```python
lc = plt.figure(figsize=(10,6))
eixos = lc.add_axes([1,1,1,1])

## aqui purple significa a cor que será usada na linha e lw é o width
eixos.plot(dates["Date"], dates["Clicks"], "purple", lw=4)
eixos.set_ylabel("Clicks", fontsize=15)
eixos.set_xlabel("Datas de acesso", fontsize=15)
eixos = plt.xticks(rotation="90")
eixos = plt.title("Clicks por data", fontsize=20)
plt.legend(['Clicks'], loc = 'lower right', fontsize=15)
```

Quando você quer plotar duas linhas diferentes no mesmo gráfico basta você adicionar na variável eixos os dados e formatá-los do jeito que você quiser. Por exemplo, se você quiser mostrar os clicks e impressões do website no mesmo gráfico basta fazer o seguinte:

```python
lc = plt.figure(figsize=(10,6))
eixos = lc.add_axes([1,1,1,1])

## aqui purple significa a cor que será usada na linha e lw é o width
eixos.plot(dates["Date"], dates["Clicks"], "purple", lw=4)
eixos.plot(dates["Date"], dates['Impressions'],"g", lw=4)
eixos.set_ylabel("Clicks", fontsize=15)
eixos.set_xlabel("Datas de acesso", fontsize=15)
eixos = plt.xticks(rotation="90")
eixos = plt.title("Impressões e clicks", fontsize=20)
plt.legend(['Clicks','Impressions'], loc = 'lower right', fontsize=15)
```

Outro bom ajuste que você pode fazer é adicionar pontos específicos do gráfico. Isso traz uma boa impressão na visualização:


```python
lc = plt.figure(figsize=(10,6))
eixos = lc.add_axes([1,1,1,1])

## aqui purple significa a cor que será usada na linha e lw é o width
eixos.plot(dates["Date"], dates["Clicks"], marker="o", linestyle="--", color="r", lw=2)
eixos.plot(dates["Date"], dates['Impressions'], marker="o", linestyle="--", color="g", lw=2)
eixos.set_ylabel("Clicks", fontsize=15)
eixos.set_xlabel("Datas de acesso", fontsize=15)
eixos = plt.xticks(rotation="90")
eixos = plt.title("Impressões e clicks", fontsize=20)
plt.legend(['Clicks','Impressions'], loc = 'lower right', fontsize=15)
```

Agora vamos para o exercício do ensino médio que é utilizar o matplotlib para plotar o gráfico de uma função do primeiro grau. Esse gráfico é definido por uma simples reta ou seja uma reta que passa por dois pontos.

Portanto vamos definir manualmente dois pontos (4,2) e (6,7) e vamos tentar criar um gráfico que mostra uma linha passando por esses dois pontos.

```python
x = [4,6]
y = [6,7]

plt.plot(x,y, linestyle="--", marker="o", lw=2, color="g")
plt.legend(["Pontos"])
```

Perfeito! agora vamos criar uma reta baseando-se apenas em uma função da reta. Para relembrar a função da reta é determinada por:
    
y(x) = ax + b

Em que "a" e "b" são pesos e você poderá alterá-los da forma que desejar e ver essa reta ser criada de acordo com sua função.

```python


def reta(x):
    a = 1
    b = 1
    return a*x + b

x = [5,7,3,8,3,5]
y = []
for n in x:
    y.append(reta(n))
    
print(y)
    
plt.xlim([-4,10])
plt.ylim([-4,10])

plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.plot(x,y,color='r', marker='o', linestyle='--')
```

Mas e se nós fizessemos o mesmo usando uma curva? famosa equação do segundo grau? Certo, para isso precisariamos modificar nossa função para mostrar uma curva. Essa modificação inclui a adição de ao menos 3 pesos (a,b,c) e também um dos termos precisa ser quadrático.

y(x) = ax² + bx + c

Ficaria mais ou menos assim:

```python
import numpy as np
from numpy import random
from scipy.interpolate import make_interp_spline, BSpline

## aqui definimos a função da curva
def curva(x):
    a = 2
    b = 4
    c = -6
    return a*x**2 + b*x + c

x = [-10,-8,-6,-4,-2,0,2,4,6,8,10]
y = []

## calcula os valores de y
for n in x:
    y.append(curva(n))
    
print(y)
    

fig = plt.figure(figsize=(10,8))
plt.xlim([-25,35])
plt.ylim([-25,35])

## plota duas linhas no eixo x e y na posição 0
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')

## faz um smooth da linha para realmente parecer uma curva
xnew = np.linspace(min(x), max(x), 200) 
spl = make_interp_spline(x, y, k=3)
y_smooth = spl(xnew)

plt.plot(xnew,y_smooth,color='r' )
```
